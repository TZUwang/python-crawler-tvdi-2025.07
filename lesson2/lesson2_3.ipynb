```python
import requests              #使用外部套件requests(連結internet,下載網頁)
from pprint import pprint    #從pprint這一odule找出pprint這一function

r = requests.get('https://api.github.com/events')
if r.status_code == 200:
    print("下載成功")
    pprint(r.text)  #要看是甚麼類型 利用print(type(r))
else:               #若上述不成立(status_code不等於200) 即執行else
    print("下載失敗")
```
